**11. Identify and remove outliers in the km_driven column using the IQR method?**

```python
import pandas as pd

# df = pd.read_csv("BIKE DETAILS.csv")

# df['Kms_Driven'] = df['Kms_Driven'].replace('[^0-9]', '', regex=True).astype(int)

Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

removed = len(df) - len(df_cleaned)
print(f"Outliers removed: {removed}")
print(f"Remaining rows: {len(df_cleaned)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("spotify.csv")

artist_col = 'Artist'
duration_col = 'Duration (ms)'

if artist_col not in df.columns or duration_col not in df.columns:
    print(f"Required columns '{artist_col}' or '{duration_col}' not found in the DataFrame.")
else:

    df_clean = df.dropna(subset=[artist_col, duration_col])

    # Step 1: Find top artists by track count (e.g., top 10)
    top_artists = df_clean[artist_col].value_counts().head(10).index.tolist()

    # Step 2: Filter data for these top artists
    top_artists_data = df_clean[df_clean[artist_col].isin(top_artists)].copy() # Use .copy() to avoid SettingWithCopyWarning

    # Convert duration to minutes for better readability
    top_artists_data['Duration_min'] = top_artists_data[duration_col] / 60000

    # Visualize the distribution of track duration for the top artists using a box plot
    plt.figure(figsize=(14, 7))
    sns.boxplot(data=top_artists_data, x=artist_col, y='Duration_min', order=top_artists)
    plt.title("Distribution of Track Duration by Top 10 Artists")
    plt.xlabel("Artist")
    plt.ylabel("Duration (minutes)")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

**10.How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("spotify.csv") # Load the data

# Determine artist and popularity column names
artist_col = 'Artist'
popularity_col = 'Popularity'

# Check if required columns exist
if artist_col not in df.columns or popularity_col not in df.columns:
    print(f"Required columns '{artist_col}' or '{popularity_col}' not found in the DataFrame.")
else:
    # Clean data: drop rows missing artist or popularity
    df_clean = df.dropna(subset=[artist_col, popularity_col])

    # Step 1: Find top artists by track count (e.g., top 10)
    top_artists = df_clean[artist_col].value_counts().head(10).index.tolist()

    # Step 2: Filter data for these top artists
    top_artists_data = df_clean[df_clean[artist_col].isin(top_artists)].copy() # Use .copy() to avoid SettingWithCopyWarning

    # Visualize the distribution of track popularity for the top artists using a violin plot
    plt.figure(figsize=(14, 7))
    sns.violinplot(data=top_artists_data, x=artist_col, y=popularity_col, order=top_artists)
    plt.title("Distribution of Track Popularity by Top 10 Artists (Violin Plot)")
    plt.xlabel("Artist")
    plt.ylabel("Popularity Score")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
